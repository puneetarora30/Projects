```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dataset = pd.read_csv("/content/covid19_Confirmed_dataset.csv")
dataset.head()
```

```python
dataset.drop(["Lat","Long","Province/State"], axis=1,inplace=True)
```

```python
dataset_country_aggregated = dataset.groupby("Country/Region").sum()
```

```python
countries = list (dataset_country_aggregated.index)
max_infection_rates = []
for c in countries:
  max_infection_rates.append(dataset_country_aggregated.loc[c].diff().max())
dataset_country_aggregated["Max_infection_rates"] = max_infection_rates


```

```python
dataset_country_aggregated
```

```python
max_corona_data = pd.DataFrame(dataset_country_aggregated["Max_infection_rates"])
max_corona_data
```
